```python
single_tool_single_arg = 0
single_tool_multiple_arg = 0
multiple_tools = 0
tool_with_no_args = 0
empty_answers = 0
ambigous_examples = 0
merged_data_without_err_answers = []
for data in new_data:
    answer = data['answer']
    if not isinstance(answer,list):
        print(answer)
        data['answer'] = [answer]
        #answer = [answer]
    
    #print(answer[0].keys())
    if(len(answer) == 0):
        empty_answers += 1
        #print(data)
        merged_data_without_err_answers.append({"question": data['question'], "answer": answer})
    elif(len(answer) == 1):
        if('arguments' not in answer[0].keys() and 'type' in answer[0].keys()):
            ambigous_examples += 1
        elif('arguments' not in answer[0].keys()):
            print(data)
        elif(len(answer[0]['arguments']) == 1):
            single_tool_single_arg += 1
            merged_data_without_err_answers.append({"question": data['question'], "answer": answer})
        elif(len(answer[0]['arguments']) == 0):
            #print(data)
            merged_data_without_err_answers.append({"question": data['question'], "answer": answer})
            tool_with_no_args += 1
        else: 
            single_tool_multiple_arg += 1
            merged_data_without_err_answers.append({"question": data['question'], "answer": answer})
    else:
        multiple_tools += 1
        merged_data_without_err_answers.append({"question": data['question'], "answer": answer})
x.add_row(["Single Tool Single Argument", single_tool_single_arg, "Merged Data"])
x.add_row(["Single Tool Multiple Argument", single_tool_multiple_arg, "Merged Data"])
x.add_row(["Multiple Tools", multiple_tools, "Merged Data"])
x.add_row(["Tool With No Arguments", tool_with_no_args, "Merged Data"])
x.add_row(["Empty Answers", empty_answers, "Merged Data"])
x.add_row(["Ambigous Examples", ambigous_examples, "Merged Data"])
x.add_row(["total data len",len(merged_data_without_err_answers), "Merged Data without error answers"])
x.add_row(["Total Tools", single_tool_single_arg + single_tool_multiple_arg + multiple_tools + empty_answers + tool_with_no_args + ambigous_examples, "Merged Data"])
x
```

```python
examples_for_zero_shot = []
for data in merged_data:
    if(len(data['answer']) == 1):
        
        if('tool_name' in list(data['answer'][0].keys())):
            #print(data['answer'][0]['tool_name'] in zero_shot_tool_list)
            if(data['answer'][0]['tool_name'] in zero_shot_tool_list):
                examples_for_zero_shot.append(data)
    elif(len(data['answer']) > 1):
        #print(data['answer'][0].keys(), len(data['answer']))
        for tool in data['answer']:
            if('tool_name' in tool.keys()):
                if(tool['tool_name'] in zero_shot_tool_list):
                    examples_for_zero_shot.append(data)
examples_for_zero_shot
```

```python
test_with_single_tool_single_args = []
test_with_single_tool_single_args_max_count = 20

test_with_single_tool_multiple_args = []
test_with_single_tool_multiple_args_max_count = 15
test_with_multiple_tools = []
test_with_multiple_tools_max_count = 20
test_with_empty_answers = []
test_with_empty_answers_max_count = 2
test_with_tools_with_no_args = []
test_with_tools_with_no_args_max_count = 2
zero_shot_test_tools = []


train_with_single_tool_single_args = []
train_with_single_tool_multiple_args = []
train_with_multiple_tools = []
train_with_empty_answers = []
train_with_tools_with_no_args = []


for data in merged_data:
    answer = data['answer']
    if(len(data['answer'])==0):
        if(len(test_with_empty_answers) < test_with_empty_answers_max_count):
            test_with_empty_answers.append(data)
        else:
            train_with_empty_answers.append(data)
    elif(len(data['answer'])==1):
        if('tool_name' in list(data['answer'][0].keys())):
            if(data['answer'][0]['tool_name'] in zero_shot_tool_list):
                zero_shot_test_tools.append(data) 
            elif(len(data['answer'][0]['arguments']) == 1):
                if(len(test_with_single_tool_single_args) < test_with_single_tool_single_args_max_count):
                    test_with_single_tool_single_args.append(data)
                else:
                    train_with_single_tool_single_args.append(data)
            elif(len(data['answer'][0]['arguments']) > 1):
                if(len(test_with_single_tool_multiple_args) < test_with_single_tool_multiple_args_max_count):
                    test_with_single_tool_multiple_args.append(data)
                else:
                    train_with_single_tool_multiple_args.append(data)
            else:
                if(len(test_with_tools_with_no_args) < test_with_tools_with_no_args_max_count):
                    test_with_tools_with_no_args.append(data)
                else:
                    train_with_tools_with_no_args.append(data)
    else:
        for tool in data['answer']:
            if(tool['tool_name'] in zero_shot_tool_list):
                zero_shot_test_tools.append(data)
        else:
            if(len(test_with_multiple_tools) < test_with_multiple_tools_max_count):
                test_with_multiple_tools.append(data)
            else:
                train_with_multiple_tools.append(data)
                
            # if(len(test_with_tools_with_no_args) < test_with_tools_with_no_args_max_count):
            #     test_with_tools_with_no_args.append(data)
            # else:
            #     train_with_tools_with_no_args.append(data)
```
